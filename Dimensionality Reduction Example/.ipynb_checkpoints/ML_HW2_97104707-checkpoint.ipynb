```python
fig, axes = plt.subplots(2,2, figsize=(8,8))

r1 = np.random.randn(50)/3
r2 = np.random.randn(50)/3
scatter_hue = np.random.choice([True, False], size=50)
r1 += np.ones(50) * ~scatter_hue * 0.5
axes[0][0].set_title('Scatter plot')
sns.scatterplot(x=r1, y=r2, hue=scatter_hue, ax=axes[0][0])

dist_smaples = list(np.random.normal(loc=0.0, scale=1, size=20)) + [8,8,-32]
axes[0][1].set_title('Histogram')
sns.distplot(dist_smaples, bins=20, hist=True, ax=axes[0][1])

pie_values = [10, 30, 2, 45]
axes[1][0].set_title('Pie Chart')
axes[1][0].pie(sorted(pie_values), labels=sorted(pie_values))

box_values = np.concatenate((2 + np.random.randn(20), -5 + 2 * np.random.randn(5)))
axes[1][1].set_title('Box Plot')
sns.boxplot(box_values, ax=axes[1][1]);
```

Looking at the plot above, there are lots of data points outside the whiskers of the box plot in columns "months" and "expenditure" so we can conclude there are outliers in these columns. Other columns like "age" or "active" seem to have some outliers too but we'll ignore those.

<div dir="rtl">
در برخی از ستون‌های اسمی عباراتی وجود دارند که معادل ناموجود بودن مقدار آن ردیف در ان ستون هستند. همچنین ستون‌های اسمی باینری نیز بهتر است به حالت True, False در بیایند.
        برای ستون‌های عددی نیز برخی مقادیر غیر مجاز یا ناموجود هستند(یعنی برای ستون‌هایی که باید مقدار مثبت داشته باشند، مقدار منفی ثبت شده است و یا داده‌ای بسیار بزرگ به عبارتی outlier وجود دارد) که باید به صورت None در بیایند.
    این ستون‌ها را تغییر دهید و دوباره نسبت مقادیر ناموجود هر ستون و همینطور اطلاعات آماری ستون های عددی تغییر یافته را محاسبه کنید.
    <br>
    (۵ نمره)
</div>

```python
# Calc missing fields

#change categorical column's type to bool
df_clean['owner'] = df_clean['owner'].map({'yes':True,'no':False, '1': True, '0': False})
df_clean['selfemp'] = df_clean['selfemp'].map({'yes':True,'no':False, 'no-info': None})

#no negative data is allowed
df_clean = df_clean[df_clean.share >= 0]
df_clean = df_clean[df_clean.expenditure >= 0]
df_clean = df_clean[df_clean.active >= 0]

#removing outliers in "expenditure" and "months"
q_low_exp = df_clean['expenditure'].quantile(0.2)
q_hi_exp  = df_clean['expenditure'].quantile(0.8)
df_clean.loc[(df_clean.expenditure > q_hi_exp) | (df_clean.expenditure < q_low_exp) , ['expenditure']] = None

q_low_mon = df_clean['months'].quantile(0.2)
q_hi_mon = df_clean['months'].quantile(0.8)
df_clean.loc[(df_clean.months > q_hi_mon) | (df_clean.months < q_low_mon) , ['months']] = None

df_clean.select_dtypes(include='number').plot.box(title="box plot", grid = True)
plt.show(block=True);
```

```python
conditions = [(0 <= df_clean.age) & (df_clean.age <= 35), (35<df_clean.age) & (df_clean.age<=55), (55<df_clean.age)]
values = [0,1,2]
df_clean['groupAge'] = np.select(conditions, values)

young = df_clean[df_clean.groupAge == 0]['income']
middle =  df_clean[df_clean.groupAge == 1]['income']
old =  df_clean[df_clean.groupAge == 2]['income']
sns.distplot(young,label = '0 to 35', hist=False)
sns.distplot(middle, label = '35 to 55', hist = False)
sns.distplot(old, label = 'older than 55', hist = False)
plt.legend()
plt.title("income distribution of different age groups")
plt.show()
```

    در ابتدا دیتا ست mnist را از فایل لود کرده و یک یا چند داده از آن را نمایش دهید.

```python
# Load and show some of data points
from PIL import Image
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
with np.load('mnist.zip') as data:
    arr, label = data['X'], data['y']
print("arr")
print(arr.shape)
print("label")
print(label.shape)
Image.fromarray(arr[0])
```

<div dir=rtl>
با دو روش pca و tsne و با استفاده از ۲ تا از مقادیر ویژه ابعاد دیتاست را کاهش دهید.
    برای روش pca مقادیر و بردارهای ویژه‌ی استفاده شده و درصد واریانس حفظ شده را مشخص کنید.
    <br>
    (۵ نمره)
</div>

```python
arr = arr.reshape(10000, 28*28)
pca = PCA(n_components=2)
pca.fit(arr)
print("Eigen Values: ", pca.explained_variance_)
print("Eigen Vectors: ", pca.components_)
print("Remained Var: ", pca.explained_variance_ratio_)
arrpca = pca.transform(arr)
tsne = TSNE(n_components = 2, random_state = 0)
arrtsne = tsne.fit_transform(arr)
```

<div dir=rtl>
دیتاست‌های حاصل را در قالب نمودار دوبعدی نمایش داده و قابلیت این دو روش را در جداسازی کلاس‌های مختلف و با استفاده از لیبل داده‌ها مقایسه کنید. علت برتری یک روش به روش دیگر را به طور خلاصه توضیح دهید.
    <br>
    (۴ نمره)
</div>

```python
colors = {0: 'blue', 1:'yellow', 2:'red', 3: 'gray', 4:'green', 5:'black', 6:'orange', 7:'pink', 8:'brown', 9:'cyan'}
def plot_data(datas):    
    
    fg, ax = plt.subplots()

    for h in np.unique(label):
        p = np.where(label == h)
        ax.scatter(datas[p, 0], datas[p, 1], c = colors[h], label = h, s = 100, alpha=0.2)
    ax.legend()
    plt.show()

print("PCA plot:")
plot_data(arrpca)
print("TSNE plot:")
plot_data(arrtsne)
```

<div dir=rtl>
    روش tsne  با توجه به نمودار بالا با دقت بسیار بهتری اعداد را از هم جدا کرده و در کل با دقت بالاتری عمل کرده.
    <br>
    اما روش pca به نسبت Tsne با سرعت بیشتری عملیات لازم را انجام می‌دهد.

<div dir=rtl>
حال میخواهیم pca را بر روی دیتاست اصلی خود بدون استفاده از کتابخانه‌ی sklearn پیاده سازی کنیم. برای اینکار در ابتدا ستون‌های فیچر را جداکرده و سپس داده‌ها را نرمالایز کنید. لزوم نرمالایز کردن داده‌ها برای روش pca به چه خصوصیتی از این روش باز میگردد؟
    <br>
    (۴ نمره)
</div>

```python
df_data = df_clean.copy()
labels = df_data['wealth'].to_numpy()
df_arr = df_data.drop(columns=['income','wealth', 'groupAge']).to_numpy().astype(float)
for i in  range(df_arr.shape[1]):
  df_arr[:, i] = (df_arr[:, i] - np.mean(df_arr[:, i])) / np.std(df_arr[:, i])

print("### MEAN: ", np.mean(df_arr))
print("### STD: ", np.std(df_arr))
```

<div dir=rtl>
حال ماتریس کوواریانس دیتاست موجود را بدست آورید و مقادیر و بردار‌های ویژه‌ی آن را به ترتیب نزولی مقادیر ویژه نمایش دهید.
    <br>
    (۳ نمره)
</div>

```python
cov_arr = np.cov(df_arr, rowvar = False)
eigen_vals, eigen_vecs = np.linalg.eigh(cov_arr)
x = eigen_vals.argsort()[::-1]   
eigen_vals = eigen_vals[x]
eigen_vecs = eigen_vecs[:,x]
print("### Sorted Eigen Values", eigen_vals)
print()
print("### Eigen Vectors", eigen_vecs)
```

<div dir=rtl>
همه‌ی
    explained variance 
    را بدست بیاورید و در نمودار pie نمایش دهید
    حتما درصد هرکدام از اسلایس‌ها در نمودار مشخص شده‌باشد.
    <br>
    (۴ نمره)
</div>

```python
explained_var = eigen_vals / sum(eigen_vals)
print(explained_var)
plt.pie(explained_var, autopct="%.1f%%", labels = [i + 1 for i in range(len(eigen_vecs))])
plt.show()
```

<div dir=rtl>
با حفظ حداقل ۵۰ درصد واریانس داده‌ها، ماتریس پرتو(projection matrix) را بدست آورید و تبدیل pca را بر روی دیتاست اجرا کنید.
    <br>
    (۵ نمره)
</div>

```python
proj_mat = eigen_vecs[:,0:3].transpose()
pca_d = np.dot(proj_mat , df_arr.transpose()).transpose()
```

<div dir=rtl>
با استفاده از مقادیر متناظر با pcهای اول و دوم هر داده در دیتاست حاصل شده، نموداری رسم کرده و تاثیر این روش را در جداسازی دو کلاس فقیر و ثروتمند تحلیل کنید.
    <br>
    (۴ نمره)
</div>

```python
color = {0: 'red', 1:'yellow'}
legend = {0: 'non weathly', 1: 'wealthy'}

fig, ax = plt.subplots()

for h in range(2):
    p = np.where(labels == h)
    ax.scatter(pca_d[p, 0], pca_d[p, 1], c = color[h], label = legend[h], s = 100, alpha=0.3)
ax.legend()
plt.show()
```

<div dir=rtl>
    با توجه به نمودار می‌بینیم که دو دسته به خوبی از هم جدا نشده‌اند و این روش نتوانسته به خوبی دو طبقه را از هم تمایز دهد.
