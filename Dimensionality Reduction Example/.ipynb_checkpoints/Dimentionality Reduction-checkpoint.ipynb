```python
import numpy as np
import pandas as pd 

import matplotlib.pyplot as plt
import seaborn as sns               
%matplotlib inline
sns.set()
plt.style.use("ggplot")

import warnings                                            
warnings.filterwarnings('ignore')
```

<div dir="ltr">
    <h2>
Dimensionality Reduction
</div>

<div dir=ltr>
    We'll be loading mnist.zip dataset and show one random data stored in it 
</div>

```python
# Load and show some of data points
from PIL import Image
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
with np.load('mnist.zip') as data:
    arr, label = data['X'], data['y']
print("arr")
print(arr.shape)
print("label")
print(label.shape)
Image.fromarray(arr[0])
```

<div dir=ltr>
Using PCA and TSNE, we will then reduce dimentionality of dataset using 2 eigen values.
</div>

```python
arr = arr.reshape(10000, 28*28)
pca = PCA(n_components=2)
pca.fit(arr)
print("Eigen Values: ", pca.explained_variance_)
print("Eigen Vectors: ", pca.components_)
print("Remained Var: ", pca.explained_variance_ratio_)
arrpca = pca.transform(arr)
tsne = TSNE(n_components = 2, random_state = 0)
arrtsne = tsne.fit_transform(arr)
```

<div dir=ltr>
Then we will show each method's functionality by plotting the results.
</div>

```python
colors = {0: 'blue', 1:'yellow', 2:'red', 3: 'gray', 4:'green', 5:'black', 6:'orange', 7:'pink', 8:'brown', 9:'cyan'}
def plot_data(datas):    
    
    fg, ax = plt.subplots()

    for h in np.unique(label):
        p = np.where(label == h)
        ax.scatter(datas[p, 0], datas[p, 1], c = colors[h], label = h, s = 100, alpha=0.2)
    ax.legend()
    plt.show()

print("PCA plot:")
plot_data(arrpca)
print("TSNE plot:")
plot_data(arrtsne)
```

<div dir=ltr>
    Looking at the plots, we can see TSNE did a better job classifying the images. However, PCA is a faster method compared to TSNE.
</div>
